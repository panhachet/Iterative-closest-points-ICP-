```python
import numpy as np
from scipy.spatial import KDTree

def nearest_neighbor(src, dst):
    """
    Find the nearest (Euclidean) neighbor in dst for each point in src.
    :param src: Source point cloud (Nx3)
    :param dst: Destination point cloud (Mx3)
    :return: Corresponding indices of the nearest points in dst
    """
    tree = KDTree(dst)
    distances, indices = tree.query(src)
    return indices

def icp(src, dst, max_iterations=100, tolerance=1e-6):
    """
    Perform ICP (Iterative Closest Point) algorithm.
    :param src: Source point cloud (Nx3)
    :param dst: Destination point cloud (Mx3)
    :param max_iterations: Maximum number of iterations
    :param tolerance: Tolerance for convergence
    :return: Final transformation matrix and aligned source point cloud
    """
    src_homogeneous = np.hstack((src, np.ones((src.shape[0], 1))))  # Convert to homogeneous coordinates
    dst_homogeneous = np.hstack((dst, np.ones((dst.shape[0], 1))))

    prev_error = 0

    for i in range(max_iterations):
        # Step 1: Find the nearest neighbors in the destination for each point in the source
        indices = nearest_neighbor(src, dst)
        dst_matched = dst[indices]

        # Step 2: Compute the centroids of both point sets
        src_centroid = np.mean(src, axis=0)
        dst_centroid = np.mean(dst_matched, axis=0)

        # Step 3: Subtract centroids from point sets
        src_centered = src - src_centroid
        dst_centered = dst_matched - dst_centroid

        # Step 4: Compute the covariance matrix H
        H = src_centered.T @ dst_centered

        # Step 5: Singular Value Decomposition (SVD)
        U, S, Vt = np.linalg.svd(H)
        R = Vt.T @ U.T

        # Ensure we have a valid rotation matrix (det(R) = 1)
        if np.linalg.det(R) < 0:
            Vt[-1, :] *= -1
            R = Vt.T @ U.T

        # Step 6: Compute the translation vector
        t = dst_centroid.T - R @ src_centroid.T

        # Step 7: Apply the transformation (rotation + translation)
        src = (R @ src.T).T + t

        # Step 8: Check for convergence
        mean_error = np.mean(np.linalg.norm(src - dst_matched, axis=1))
        if np.abs(prev_error - mean_error) < tolerance:
            break
        prev_error = mean_error

    # Final transformation matrix
    transformation = np.eye(4)
    transformation[:3, :3] = R
    transformation[:3, 3] = t

    return transformation, src

# Example of two arrays of points
src_points = np.array([[1.0, 1.0, 0.0],
                       [2.0, 1.0, 0.0],
                       [3.0, 1.0, 0.0],
                       [4.0, 1.0, 0.0]])

dst_points = np.array([[1.1, 1.2, 0.0],
                       [2.1, 1.1, 0.0],
                       [3.0, 1.0, 0.0],
                       [4.2, 1.0, 0.0]])

# Perform ICP
transformation_matrix, aligned_src = icp(src_points, dst_points)

# Output the results
print("Transformation Matrix:")
print(transformation_matrix)

print("\nAligned Source Points:")
print(aligned_src)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def nearest_neighbor_1d(src, dst):
    """
    Find the nearest neighbor in dst for each point in src (1D case).
    :param src: Source array (N,)
    :param dst: Destination array (M,)
    :return: Corresponding indices of the nearest points in dst
    """
    indices = np.argmin(np.abs(dst[:, np.newaxis] - src), axis=0)
    return indices

def icp_1d(src, dst, max_iterations=100, tolerance=1e-6):
    """
    Perform ICP for 1D arrays (iterative closest point).
    :param src: Source array (N,)
    :param dst: Destination array (M,)
    :param max_iterations: Maximum number of iterations
    :param tolerance: Tolerance for convergence
    :return: Translation and the aligned source array
    """
    prev_error = 0

    for i in range(max_iterations):
        # Step 1: Find the nearest neighbors in the destination for each point in the source
        indices = nearest_neighbor_1d(src, dst)
        dst_matched = dst[indices]

        # Step 2: Compute the centroids of both point sets
        src_centroid = np.mean(src)
        dst_centroid = np.mean(dst_matched)

        # Step 3: Compute the translation required
        translation = dst_centroid - src_centroid

        # Step 4: Apply the translation to the source array
        src = src + translation

        # Step 5: Check for convergence
        mean_error = np.mean(np.abs(src - dst_matched))
        if np.abs(prev_error - mean_error) < tolerance:
            break
        prev_error = mean_error

    return translation, src

# Example 1D arrays
src_points = np.array([1.0, 2.0, 3.0, 4.0])
dst_points = np.array([1.5, 2.5, 3.5, 4.5])

# Perform ICP for 1D arrays
translation, aligned_src = icp_1d(src_points, dst_points)

# Plot the original and aligned points
plt.figure()

# Plot original source points (before alignment)
plt.scatter(src_points, np.zeros_like(src_points), color='red', label='Source Points', marker='o')

# Plot destination points
plt.scatter(dst_points, np.zeros_like(dst_points), color='green', label='Destination Points', marker='x')

# Plot aligned source points (after ICP)
plt.scatter(aligned_src, np.zeros_like(aligned_src), color='blue', label='Aligned Source Points', marker='^')

# Add labels and title
plt.title('ICP for 1D Arrays')
plt.xlabel('1D Points')
plt.yticks([])  # Hide Y-axis since it's 1D
plt.legend()

# Show the plot
plt.show()

# Output the results
print("Translation:", translation)
print("Aligned Source Points:", aligned_src)
```

```python
import numpy as np
from scipy.spatial import KDTree
import matplotlib.pyplot as plt

def nearest_neighbor_2d(Q, P):
    """
    Find the nearest neighbor in P for each point in Q (2D case).
    :param Q: Source point cloud (Nx2)
    :param P: Destination point cloud (Mx2)
    :return: Corresponding indices of the nearest points in P
    """
    tree = KDTree(P)
    distances, indices = tree.query(Q)
    return indices

def icp_2d(Q, P, max_iterations=100, tolerance=1e-6):
    """
    Perform ICP for 2D arrays (iterative closest point).
    :param Q: Source point cloud (Nx2)
    :param P: Destination point cloud (Mx2)
    :param max_iterations: Maximum number of iterations
    :param tolerance: Tolerance for convergence
    :return: Transformation matrix and aligned source points
    """
    prev_error = 0

    for i in range(max_iterations):
        # Step 1: Find the nearest neighbors in the destination for each point in the source
        indices = nearest_neighbor_2d(Q, P)
        P_matched = P[indices]

        # Step 2: Compute the centroids of both point sets
        Q_centroid = np.mean(Q, axis=0)
        P_centroid = np.mean(P_matched, axis=0)

        # Step 3: Subtract the centroids from the points
        Q_centered = Q - Q_centroid
        P_centered = P_matched - P_centroid

        # Step 4: Compute the covariance matrix
        H = Q_centered.T @ P_centered

        # Step 5: Perform Singular Value Decomposition (SVD)
        U, S, Vt = np.linalg.svd(H)
        R = Vt.T @ U.T

        # Ensure we have a valid rotation matrix (det(R) = 1)
        if np.linalg.det(R) < 0:
            Vt[-1, :] *= -1
            R = Vt.T @ U.T

        # Step 6: Compute the translation vector
        t = P_centroid - R @ Q_centroid

        # Step 7: Apply the transformation (rotation + translation)
        Q = (R @ Q.T).T + t

        # Step 8: Check for convergence
        mean_error = np.mean(np.linalg.norm(Q - P_matched, axis=1))
        if np.abs(prev_error - mean_error) < tolerance:
            break
        prev_error = mean_error

    # Return the transformation matrix and aligned source points
    transformation_matrix = np.eye(3)
    transformation_matrix[:2, :2] = R
    transformation_matrix[:2, 2] = t

    return transformation_matrix, Q

# Example 2D points
Qx = np.array([1.0, 2.0, 3.0, 4.0])
Qy = np.array([1.0, 1.0, 1.0, 1.0])
Q = np.column_stack((Qx, Qy))

Px = np.array([1.1, 2.1, 3.0, 4.2])
Py = np.array([1.2, 1.1, 1.0, 1.0])
P = np.column_stack((Px, Py))

# Perform ICP for 2D points
transformation_matrix, aligned_Q = icp_2d(Q, P)

# Plot the original and aligned points
plt.figure()

# Plot original source points (before alignment)
plt.scatter(Q[:, 0], Q[:, 1], color='red', label='Source Points (Q)', marker='o')

# Plot destination points
plt.scatter(P[:, 0], P[:, 1], color='green', label='Destination Points (P)', marker='x')

# Plot aligned source points (after ICP)
plt.scatter(aligned_Q[:, 0], aligned_Q[:, 1], color='blue', label='Aligned Source Points (Q)', marker='^')

# Add labels and title
plt.title('ICP for 2D Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Show the plot
plt.show()

# Output the results
print("Transformation Matrix:")
print(transformation_matrix)

print("\nAligned Source Points:")
print(aligned_Q)
```

```python
import numpy as np

def center_data(data, exclude_indices=[]):
    # Remove the columns based on exclude_indices (exclude specific points)
    reduced_data = np.delete(data, exclude_indices, axis=1)
    
    # Compute the mean (center) along each row (feature-wise mean)
    center = np.mean(reduced_data, axis=1, keepdims=True)
    
    # Center the original data by subtracting the mean
    centered_data = data - center
    
    return center, centered_data

# Example usage
data = np.array([1.0, 2.0, 3.0, 4.0],
                 )

exclude_indices = [2]  # Exclude the third column (index 2)

center, centered_data = center_data(data, exclude_indices)
print("Center:\n", center)
print("Centered Data:\n", centered_data)
```
