```python


def quaternion_to_angle(qx, qy, qz, qw):

    siny_cosp = 2 * (qw * qz + qx * qy)
    cosy_cosp = 1 - 2 * (qy**2 + qz**2)
    yaw = np.arctan2(siny_cosp, cosy_cosp)
    return yaw

initial_position = [-1.999929215465731, -0.5000010159091588]
initial_orientation = [4.93512682062894e-05, 0.0028678706643788555, 2.9672828704660374e-06, 0.9999958864282894]

final_position = [-1.6570323177584991, -0.3833609535278473]
final_orientation = [-0.0015148509261806362, 0.0024362987198474077, 0.5422423413310761, 0.8402172415173425]

initial_yaw = quaternion_to_angle(*initial_orientation)
final_yaw = quaternion_to_angle(*final_orientation)


plt.figure(figsize=(8, 8))
plt.quiver(initial_position[0], initial_position[1], np.cos(initial_yaw), np.sin(initial_yaw), color='r', scale=5, label='Initial Position and Orientation')
plt.quiver(final_position[0], final_position[1], np.cos(final_yaw), np.sin(final_yaw), color='b', scale=5, label='Final Position and Orientation')


plt.scatter(*initial_position, color='red', label='Initial Position', zorder=5)
plt.scatter(*final_position, color='blue', label='Final Position', zorder=5)
plt.title("Initial and Final Position with Orientation (Ignoring z)")
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()

plt.xlim(-2.5, -1.5)
plt.ylim(-1, 0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
