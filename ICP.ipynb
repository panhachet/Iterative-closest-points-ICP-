# ICP Algorithm
- Step 1: Compute Correspondences 
    - Search the closest points from P to Q
    - Save the result (index, closest index, distance)
- Step 2: Compute Cross-Covariance and Mean Values
    - Compute cross-covariance matrix $K$
    $$ K = \sum_{i=1}^N p_iq_i^T = \begin{bmatrix} cov(P_x,Q_x) & cov(P_x, Q_y) \\ cov(P_y,Q_x) & cov(P_y,Q_y)\end{bmatrix}$$
    - Compute mean values of each correspondence dataset $P, Q$
    $$\bar{P} = \frac{1}{N}\sum_{i=1}^Np_i$$
    $$\bar{Q} = \frac{1}{N}\sum_{i=1}^Nq_i$$
    
    where $P = \{p_1,p_2,\dots,p_N\}$ and $Q = \{q_1,q_2,\dots,q_N\}$
    
- Step 3: Find the Rotation and Translation Value (R,T) and update data
    - Singular value decomposition to find $R$ from $k$
    $$SVD(k) = USV^T$$
    $$R = UV^T$$
    $$T = \bar{Q} - R\bar{P}$$
    $$P_{new} = RP_{old} + T$$
- Step 4: Compute error and repeat previous steps
$$err = |\bar{Q} - R\bar{P}| $$

```python
# Make Data centering
def center_data(data, exclude_indices = []):
    reduced_data = np.delete(data, exclude_indices, axis=1)
    center = np.array([reduced_data.mean(axis=1)]).T
    centered_data = data - center
    return center, centered_data

center_P, centered_P = center_data(P)
center_Q, centered_Q = center_data(Q)

plot_data(centered_P, centered_Q)
```

#  Step 1: Compute Correspondences

```python
def compute_correspondences(P, Q):
    p_size, q_size = P.shape[1], Q.shape[1]
    correspondences = []
    for i in range(p_size):
        p_point = P[:,i]
        min_dist = sys.maxsize
        chosen_idx = -1
        for j in range(q_size):
            q_point = Q[:, j]
            dist = np.linalg.norm(q_point - p_point)
            if dist < min_dist:
                min_dist = dist
                chosen_idx = j
        correspondences.append((i,chosen_idx))
    return correspondences

def plot_correspondences(P, Q, correspondences):
    for i, j in correspondences:
        x = [P[0, i], Q[0, j]]
        y = [P[1, i], Q[1, j]]
        plt.plot(x, y, color="grey")
```

```python
correspondences = compute_correspondences(centered_P, centered_Q)
plot_data(centered_P, centered_Q)
plot_correspondences(centered_P, centered_Q, correspondences)
```

# Step 2: Compute Cross-Covariance and Mean Values

```python
def compute_cross_covariance(P, Q, correspondences, kernel=lambda diff: 1.0):
    cov = np.zeros((2, 2))
    exclude_indices = []
    for i, j in correspondences:
        p_point = P[:, [i]]
        q_point = Q[:, [j]]
        weight = kernel(p_point - q_point)
        if weight < 0.01: exclude_indices.append(i)
        cov += weight * q_point.dot(p_point.T)
    return cov, exclude_indices
        
cov, _ = compute_cross_covariance(centered_P, centered_Q, correspondences)
print(cov)
```

# Find R and t from SVD decomposition

```python
U, S, V_T = np.linalg.svd(cov)
print(S)
R_found = U.dot(V_T)
t_found = center_Q - R_found.dot(center_P)
print("R_found =\n", R_found)
print("t_found =\n", t_found)
```

```python
def icp(P, Q, iteration, kernel = lambda diff: 1.0):
    center_of_Q, Q_centered = center_data(Q)
    norm_values = []
    P_values = [P.copy()]
    P_copy = P.copy()
    corresp_values = []
    exclude_indices = []
    for i in range(iteration):
        center_of_P, P_centered = center_data(P_copy, exclude_indices=exclude_indices)
        correspondences = compute_correspondences(P_centered, Q_centered)
        corresp_values.append(correspondences)
        norm_values.append(np.linalg.norm(P_centered - Q_centered))
        cov, exclude_indices = compute_cross_covariance(P_centered, Q_centered, correspondences, kernel)
        U, S, V_T = np.linalg.svd(cov)
        R = U.dot(V_T)  
        t = center_of_Q - R.dot(center_of_P)
        P_copy = R.dot(P_copy) + t
        P_values.append(P_copy)
    corresp_values.append(corresp_values[-1])
    return P_values, norm_values, corresp_values

P_values, norm_values, corresp_values = icp(P, Q, iteration=10)
plot_data(P_values[-1], Q)
```
